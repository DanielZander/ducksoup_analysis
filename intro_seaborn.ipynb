```python
import seaborn as sns
import pandas as pd
```

```python
all_data_df = pd.read_csv("data/behavior/clean_data.csv")
```

```python
all_data_df.head()
```

# Trial payoff

```python
df_analysis = all_data_df.groupby(["role", "manipulation", "player"]).mean().reset_index()
```

```python
sns.catplot(x = "role", hue="manipulation", y="trial_payoff", data= df_analysis)
```

```python

df_analysis = all_data_df.groupby(["role", "manipulation", "player"]).mean().reset_index()
df_analysis = df_analysis.loc[df_analysis["player"] != 4]

sns.catplot(x = "role", hue="manipulation", y="trial_payoff", data= df_analysis, kind="point", capsize=0.2)
```

```python

df_analysis = all_data_df.groupby(["role", "manipulation", "player"]).mean().reset_index()
df_analysis = df_analysis.loc[df_analysis["player"] != 4]

sns.catplot(x = "role", hue="manipulation", y="trial_payoff", data= df_analysis, kind="bar", capsize=0.2)
```

```python

df_analysis = all_data_df.groupby(["role", "manipulation", "player"]).mean().reset_index()
df_analysis = df_analysis.loc[df_analysis["player"] != 4]
df_analysis = df_analysis.loc[df_analysis["player"] != 5]

sns.catplot(x = "role", hue="manipulation", y="trial_payoff", data= df_analysis, kind="bar", capsize=0.2)
```

```python

sns.catplot(x = "role", hue="manipulation", y="trial_payoff", col="player", data= all_data_df, kind="bar", capsize=0.2)
```

# Personality measures

```python
df_analysis = all_data_df.groupby(["role", "player"]).mean().reset_index()

sns.lmplot(x = "mean_aggresive_dominance", y= "mean_social_dominance", data=df_analysis)
```

```python
df_analysis = all_data_df.groupby(["role", "player"]).mean().reset_index()

sns.lmplot(x = "mean_aggresive_dominance", y= "trial_payoff", hue="role", data=df_analysis)
```

```python
df_analysis = all_data_df.groupby(["role", "player"]).mean().reset_index()

sns.lmplot(x = "mean_social_dominance", y= "trial_payoff", hue="role", data=df_analysis)
```

```python
df_analysis = all_data_df.groupby(["role", "player", "manipulation"]).mean().reset_index()

sns.lmplot(x = "mean_social_dominance", y= "trial_payoff", hue="manipulation", col="role", data=df_analysis)
```
