# SEABORN INTRO GUIDE

```python
import seaborn as sns
import pandas as pd
```

```python
all_data_df = pd.read_csv("raw_data/clean_data.csv")
all_data_df.head()
```

```python
sns.histplot(all_data_df['trial_payoff'])
```

```python
df_analysis = all_data_df.groupby(["role", "manipulation", "player"])['trial_payoff'].mean().reset_index()
df_analysis
```

```python
sns.catplot(x = "role", hue="manipulation", y="trial_payoff", data= df_analysis)
```

```python
sns.barplot(x = "role", hue="manipulation", y="trial_payoff", data= df_analysis)
```

```python
df_analysis = all_data_df.groupby(["round_nb", "player", "manipulation"])['trial_payoff'].mean().reset_index()
df_analysis
```

```python
ax = sns.catplot(x = "round_nb", y = "trial_payoff", kind = "bar", col = "manipulation", data = df_analysis)
ax.set(ylim=(0,10))
```

```python
df_analysis = all_data_df.groupby(["player","role", "manipulation"])['trial_payoff'].mean().reset_index()
df_analysis = df_analysis.loc[df_analysis["player"] != 4]
```

```python
sns.catplot(x = "role", hue="manipulation", y="trial_payoff", kind = "point", dodge= True, data= df_analysis)
```
